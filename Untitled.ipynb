```python
#data = data.drop('Alley', 1)
columns = ["FireplaceQu", "PoolQC", "Fence", "MiscFeature"]
data = data.drop(columns, 1)
pd.isnull(data).sum()
```

```python
#getting relevant features

cols = ["OverallQual", "YearBuilt", "TotalBsmtSF", "GrLivArea", "SalePrice"]
subdata = data.loc[:1460, cols]
subdata
```

```python
subdata.dtypes
```

```python
#subdata["OverallQual", "YearBuilt"].min()
```

```python
features = ["OverallQual", "YearBuilt", "TotalBsmtSF", "GrLivArea", "SalePrice"]

def findMin(features):
    print("Finding the min")
    for feature in features:
        print("the min for", feature,":" , subdata[feature].min())
        

def findMax(features):
    print("Finding the max")
    for feature in features:
        print("the max for", feature,":" , subdata[feature].max())
        
def findMean(features):
    print("Finding the mean")
    for feature in features:
        print("the mean for", feature,":" , subdata[feature].mean())
        
def findMedian(features):
    print("Finding the median")
    for feature in features:
        print("the median for", feature,":" , subdata[feature].median())
        
def findStd(features):
    print("Finding the standard deviation")
    for feature in features:
        print("the standard deviation", feature,":", subdata[feature].std())

        
        
findMin(features)
print()
findMax(features)
print()
findMean(features)
print()
findMedian(features)
print()
findStd(features)
```

```python
fig = plt.figure(figsize = (20,15))

ax1 = fig.add_subplot(2,3,1)
ax1.hist(subdata['OverallQual'])
ax1.set_title("OverallQual")
ax1.set_ylabel('frequencies')
ax1.set_xlabel("rating")


ax2 = fig.add_subplot(2,3,2)
ax2.hist(subdata['YearBuilt'])
ax2.set_title("YearBuilt")
ax2.set_xlabel("year")

ax3 = fig.add_subplot(2,3,3)
ax3.hist(subdata['TotalBsmtSF'])
ax3.set_title("TotalBsmtSF")

ax4 = fig.add_subplot(2,3,4)
ax4.hist(subdata['GrLivArea'])
ax4.set_title("GrLivArea")


ax5 = fig.add_subplot(2,3,5)
ax5.hist(subdata['SalePrice'])
ax5.set_title("SalePrice")
ax5.set_xlabel("price range")

```
